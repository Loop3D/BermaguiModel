```python
from LoopStructural import GeologicalModel
from LoopStructural.visualisation import LavaVuModelViewer
from LoopStructural.visualisation import RotationAnglePlotter

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


origin = (0.0, 0.0,0.0)          
maximum = (12000,2000,13000) #8000)
bb = np.zeros((2,3))
bb[0,0] = origin[0] 
bb[0,1] = origin[1] 
bb[0,2] = origin[2]
bb[1,0] = maximum[0]
bb[1,1] = maximum[1]
bb[1,2] = maximum[2]

images = [] #to store the array containing the images for making figures
```

```python
data = pd.read_csv('./data/Bermagui_data_vectors_v2.csv')
data
```

# Modelling S5


1. Create a geological model using the bounding box
2. Associate the data to the model
3. Add the fold frame for F5 to the model 
4. Create an interactive view of the model S5 scalar field+data+isosurfaces
4. Create an image of the model at a specific view

```python
model = GeologicalModel(bb[0,:],bb[1,:])
model.set_model_data(data)
S5 = model.create_and_add_fold_frame('S5',
                                     nelements=1e4, 
                                     buffer=0.5,
                                     solver='lu',
                                     damp=True)
viewer = LavaVuModelViewer(model)
viewer.add_scalar_field(S5[0],
                       cmap='prism')
viewer.add_isosurface(S5[0],
                     nslices=5)
viewer.add_data(S5[0], vectors =True, disks = False)
rotation = [-69.11979675292969, 15.704944610595703, 6.00014591217041]
viewer.rotate(rotation)
viewer.interactive()
images.append(viewer.image_array())
#model.data
```

# Modelling S4


```python
S4 = model.create_and_add_folded_fold_frame('S4',
                                            limb_wl= 16000, 
                                            fold_frame=S5,
                                            av_fold_axis=True,
                                            nelements=1e4,
                                            buffer=0.3,
                                            lag=0.05, #svriogram lag
                                            solver='lu'
                                            )
model.update()
plt.rcParams.update({'font.size': 22})

S4.fold.foldframe[0].set_model(model)

S5_S4_splot = RotationAnglePlotter(S4)
S5_S4_splot.add_fold_limb_data()
S5_S4_splot.add_fold_limb_curve() 
S5_S4_splot.add_limb_svariogram()#fold_limb_
S5_S4_splot.ax[1,0].set_title('A. F5 Fold Limb S-Plot')
S5_S4_splot.ax[1,1].set_title('B. F5 Fold Limb S-Variogram')

S5_S4_splot.fig.delaxes(S5_S4_splot.ax[0][0])
S5_S4_splot.fig.delaxes(S5_S4_splot.ax[0][1])

from LoopStructural.modelling.features import LambdaGeologicalFeature
x_grid = LambdaGeologicalFeature(lambda XYZ: XYZ[:,0], name='X' )
z_grid = LambdaGeologicalFeature(lambda XYZ: XYZ[:,2], name='Z' )
```

```python
viewer = LavaVuModelViewer(model)
viewer.clear()
viewer.add_scalar_field(S4[0],cmap='prism')
viewer.add_isosurface(S4[0], nslices=10)
#viewer.add_isosurface(S4,nslices=5,paint_with=S4)
#viewer.add_isosurface(S4[0], nslices=10, paint_with = S4)
viewer.add_isosurface(S5[0], nslices=5, paint_with=S5[0])
#viewer.add_isosurface(S5[0], nslices=3)
viewer.add_data(S4[0], vectors =True, disks = False, color='red')
#viewer.add_data(S5[0], vectors =True, disks = False)

#rotation = [-69.11979675292969, 15.704944610595703, 6.00014591217041]
rotation = [-91.0, -2.0, 0.0]
viewer.rotate(rotation)
images.append(viewer.image_array())
viewer.interactive()
```

# Modelling S3




```python
fold_weights = {'fold_orientation':1.0, 'fold_axis_w':1.0, 'fold_regularisation':[0.1, 0.01, 0.01], 'fold_normalisation':0.25, 'fold_norm':2.5, 'step':2}
#ALL about the folded foliation // fold_orientation & fold_axis = // to folded foliation / ortho constraints;  fold_regularisation = //to fold frame;  fold_normalisation = weighting of the norm constraint in the axial surface direction/
# norm of scalar field =    // step = how many elements

model.data = pd.read_csv('./data/Bermagui_data_vectors_S3.csv')

S3 = model.create_and_add_folded_fold_frame('S3',
                                            limb_wl= 6000, #5000
                                            fold_frame=S4,
                                            fold_weights = fold_weights,
                                            #av_fold_axis=True,
                                            fold_axis=[0.,1.,0.],
                                            nelements=1e5,
                                            buffer=0.2,
                                            lag=0.05, #svriogram lag
                                            solver='lu'
                                           )
model.update()
plt.rcParams.update({'font.size': 22})

S3.fold.foldframe[0].set_model(model)

S4_S3_splot = RotationAnglePlotter(S3)
S4_S3_splot.add_fold_limb_data()
S4_S3_splot.add_fold_limb_curve() 
S4_S3_splot.add_limb_svariogram()#fold_limb_
S4_S3_splot.ax[1,0].set_title('A. F4 Fold Limb S-Plot')
S4_S3_splot.ax[1,1].set_title('B. F4 Fold Limb S-Variogram')

S4_S3_splot.fig.delaxes(S4_S3_splot.ax[0][0])
S4_S3_splot.fig.delaxes(S4_S3_splot.ax[0][1])
```

```python
viewer = LavaVuModelViewer(model)
viewer.clear()
rotation = [-87.39886474609375, 6.066004753112793, 3.7718043327331543]
viewer.rotate(rotation)
viewer.add_data(S3[0],cmap='tab20', vectors = True, disks=False)
viewer.add_isosurface(S3[0],nslices=12,paint_with=S3[0])
viewer.add_isosurface(S4[0],nslices=8, color='green') #paint_with=S4[0])

viewer.add_scalar_field(S3[0], cmap='prism')
images.append(viewer.image_array())
# for i in range(10):
#     viewer.add_isosurface(x_grid, isovalue = (i+1) * 1000)
#     viewer.add_isosurface(z_grid, isovalue = (i+1) * 1000)

viewer.interactive()

#viewer.add_data(S4[0], vectors =True, disks = False)
#viewer.add_isosurface(S5[0],nslices=3, color='yellow') #paint_with=S4[0])
#viewer.add_isosurface(S4[0], isovalue = -2000, colour='yellow')
#viewer.add_isosurface(S5[0], isovalue = -1.6, colour='yellow')
#viewer.add_points(data.loc[data['feature_name']=='S2',['X','Y','Z']]  )
#viewer.add_vector_field(s1[0],locations=model.regular_grid()[::10,:])#s0.interpolator.get_gradient_constraints()[:,:3])
#viewer.add_vector_field(s0.fold,locations=s0.interpolator.get_gradient_constraints()[:,:3])
#viewer.add_fold(S3.fold,locations=S3[0].interpolator.get_gradient_constraints()[:,:3])
#viewer.add_vector(data.loc[data['feature_name']=='S2',['X','Y','Z']], data.loc[data['feature_name']=='S2',['nx','ny','nz']],name='s2')
#viewer.add_fold(S2.fold,locations=S2[0].interpolator.get_gradient_constraints()[:,:3])
#viewer.rotate([rotation])
#=viewer.add_fold(S3.fold, [XYZ array of where to plot])    <- model_underscore_grid  model.regular_grid()[::200]
#viewer.add_fold(S3.fold, locations = model.regular_grid()[::200])      #add S3 constraints orientation as vectors
```
